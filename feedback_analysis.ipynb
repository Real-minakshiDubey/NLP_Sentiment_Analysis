```python
import nltk
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
feedback="The teacher explains concepts very clearly and the class is interesting"
```

```python
tokens=word_tokenize(feedback.lower())
print(tokens)
```

```python
import nltk
nltk.download('stopwords')
```

```python
stop_words=set(stopwords.words("english"))
filtered_words= [ w for w in tokens if w not in stop_words and w not in string.punctuation]
print(filtered_words)
```

```python
import nltk
nltk.download('averaged_perceptron_tagger')
```

```python
pos_tags=nltk.pos_tag(filtered_words)
print(pos_tags)
```

```python
import nltk
nltk.download('vader_lexicon')
```

```python
sia=SentimentIntensityAnalyzer()
sentiment_score=sia.polarity_scores(feedback)
print(sentiment_score)
```
